```python
#k-means clustering algorithm
df.columns
```

```python
X=df[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
X
#X is data frame
```

```python
#now import k-means clustering algorithm module
from sklearn.cluster import KMeans
```

```python
k_means=KMeans()
```

```python
#fit method only trained our algorithm
k_means.fit(X)
```

```python
#fit_predict not only trained but also create dependent variable that is clusters
k_means.fit_predict(X)
```

```python
#elbow method to find optimal number of clusters
wcss=[]
for i in range(1,11):
    k_means=KMeans(n_clusters=i)
    k_means.fit(X)
    wcss.append(k_means.inertia_)
    
```

```python
plt.plot(range(1,11),wcss)
plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
#in this graph optimal number is 5 
```

```python
#model training
X=df[['Annual Income (k$)', 'Spending Score (1-100)']]
k_means=KMeans(n_clusters=5,random_state=42)
y_means=k_means.fit_predict(X)
y_means
               
```

```python
plt.scatter(X.iloc[y_means==0,0],X.iloc[y_means==0,1],s=100,c='red',label='Cluster 1')
plt.scatter(X.iloc[y_means==1,0],X.iloc[y_means==1,1],s=100,c='green',label='Cluster 2')
plt.scatter(X.iloc[y_means==2,0],X.iloc[y_means==2,1],s=100,c='blue',label='Cluster 3')
plt.scatter(X.iloc[y_means==3,0],X.iloc[y_means==3,1],s=100,c='black',label='Cluster 4')
plt.scatter(X.iloc[y_means==4,0],X.iloc[y_means==4,1],s=100,c='orange',label='Cluster 5')
#now creaate cluster centroids
plt.scatter(k_means.cluster_centers_[:,0],k_means.cluster_centers_[:,1],s=100,c='magenta',label="Centroids")
plt.title("Cutomer Segmentation")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.legend()
plt.show()

```
